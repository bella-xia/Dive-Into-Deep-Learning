```python
%matplotlib inline
import torch
from d2l import torch as d2l
```

```python
eta = 0.4
def f_2d(x1, x2):
    return 0.1 * x1 ** 2 + 2 * x2 ** 2

def gd_2d(x1, x2, s1, s2):
    return (x1 - eta * 0.2 * x1, x2 - eta * 4 * x2, 0, 0)
```

```python
d2l.show_trace_2d(f_2d, d2l.train_2d(gd_2d))
```

```python
eta = 0.6
d2l.show_trace_2d(f_2d, d2l.train_2d(gd_2d))
```
